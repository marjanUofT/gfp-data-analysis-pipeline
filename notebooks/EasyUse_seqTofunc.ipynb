```python
import pandas as pd
import os

def sample_datasets(dataset_paths, n_samples, random_seed):
    """
    Samples the specified number of rows from each dataset and saves the output to CSV files.
    Ensures that the first three columns in the output are "ID", "SEQ", and "quantitative_function".

    Parameters:
    - dataset_paths: A list of tuples containing (file_path, output_file_name)
    - n_samples: Number of samples to draw from each dataset
    - random_seed: Random seed for reproducibility
    """

    # Define the target first three columns
    required_columns = ["ID", "SEQ", "quantitative_function"]

    # Generate suffix based on sample size
    sample_suffix = f"{int(n_samples/1000)}ksampled"

    # Loop through the dataset paths
    for file_path, output_file_name in dataset_paths:
        # Load the dataset
        df = pd.read_csv(file_path)

        # Check and add missing columns with default values if necessary
        if "ID" not in df.columns:
            df.insert(0, "ID", range(1, len(df) + 1))  # Add ID column as the first column
            print("there is no ID column as first column.")
        if "SEQ" not in df.columns:
            df.insert(1, "SEQ", "")  # Add SEQ column if missing with empty strings as placeholders
            print("there is no SEQ column as second column.")

        if "quantitative_function" not in df.columns:
            df["quantitative_function"] = 0  # Add quantitative_function with a default value of 0
            print("there is no quantitative_function column as third column.")

        # Reorder columns to ensure the first three are "ID", "SEQ", and "quantitative_function"
        # Keep the remaining columns in their original order after these three
        other_columns = [col for col in df.columns if col not in required_columns]
        df = df[required_columns + other_columns]

        # Sample n_rows from the dataset
        sampled_data = df.sample(n=n_samples, random_state=random_seed)

        # Get the directory of the original file to save in the same folder
        directory = os.path.dirname(file_path)

        # Create the output file name based on the sample size
        output_path = f"{output_file_name}_{sample_suffix}.csv"

        # Save the sampled data to a new CSV file in the same folder
        sampled_data.to_csv(os.path.join(directory, output_path), index=False)

    print('Sampled files have been saved.')
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

def plot_line_histograms(file_list, prpty_select, fig_path, metric = 1, metric_name=None):
    plt.figure(figsize=(10, 6))

    # Use different colors and line styles
    colors = plt.cm.get_cmap('tab10', len(file_list))


    for idx, file in enumerate(file_list):
        data = pd.read_csv(file)[prpty_select]

        # Count how many values are greater than the max value from the first file
        count_greater_than_metric = (data > metric).sum()

        # Calculate the percentage of values greater than the max value from the first file
        percentage_greater_than_metric = (count_greater_than_metric / len(data)) * 100


        # Extract the base name for the label
        name = '_'.join(os.path.splitext(os.path.basename(file))[0].split('_')[:4])

        # Plot as line plot for clarity
        plt.hist(data, bins=30, alpha=0.7, label=f'{name} (>{metric_name} = {metric}: {percentage_greater_than_metric:.2f}%)',
                 color=colors(idx), histtype='step', linewidth=2)

    plt.xlabel(prpty_select)
    plt.ylabel('Frequency')
    plt.title(f'Histogram Comparison of {prpty_select} Across Datasets')

    # Automatically adjust the x-axis limits to fit the entire data range
    plt.xlim(left=None, right=None)  # Automatically adjusts to the data's range

    plt.legend(loc='upper right')
    plt.savefig(fig_path, dpi=300)
    print("fig file is saved as: " , fig_path)
    plt.show()
```
